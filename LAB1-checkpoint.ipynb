```python
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import scipy.special as sc

#uniform distibution
uniform_distribution = np.random.uniform(0, 1, 1000) 
#normal distibution
normal_distibution_u0_s1 = np.random.normal(0, 1, 1000) 
normal_distibution_u0_s10 = np.random.normal(0, 10, 1000)

beta_distribution_a1_b1 = np.random.beta(1,1,1000) 
beta_distribution_a2_b2 = np.random.beta(2,2,1000) 
beta_distribution_a3_b9 = np.random.beta(3,9,1000)

mean = [0,0] 
E1 = [[1,0],[0,1]] 
E2 = [[2,1],[1,2]]
x1,y1 = np.random.multivariate_normal(mean,E1,10000).T 
x2,y2 = np.random.multivariate_normal(mean,E2,10000).T

```

```python
#Comparing histograms with plots of analytical formulas
#uniform distribution

a, bins, patches = plt.hist(uniform_distribution, density = True) 
y = (1/(1-0))
plt1 = plt.hlines(y, xmin=0, xmax=1, color = 'r')
plt.show()
```

```python
#Normal distribution with mean = 0 and standard deviation = 1 
n1, bins1, patches1 = plt.hist(normal_distibution_u0_s1, density = True)
y = ((1/(np.sqrt(2*np.pi)*1))*np.exp(-(bins1-0)**2/(2*1**2)));

plt2 = plt.plot(bins1,y)
plt.show()
```

```python
#normal distribution with mean = 0 and standard deviation = 10
n2, bins2, patches2 = plt.hist(normal_distibution_u0_s10, density = True)

y = ((1/(np.sqrt(2*np.pi)*10))*np.exp(-(bins2-0)**2/(2*10**2)));
plt3 = plt.plot(bins2, y, color = 'r')
plt.show()
```

```python
#beta distribution 1 -> alpha2 = 1, beta2 = 1
n3, bins3, patches3 = plt.hist(beta_distribution_a1_b1, density = True)
y = (((bins3**(1-1))*(1-bins3)**(1-1))/sc.beta(1,1));
plt4 = plt.plot(bins3, y, color = 'r')
plt.show()
```

```python
#beta distribution 2 -> alpha2 = 2, beta2 = 2
n4, bins4, patches4 = plt.hist(beta_distribution_a2_b2, density = True)
y = (((bins4**(2-1))*(1-bins4)**(2-1))/sc.beta(2,2));
plt5 = plt.plot(bins4, y, color = 'r')
plt.show()
```

```python
#beta distribution 3 -> alpha2 = 3, beta2 = 9
n5, bins5, patches6 = plt.hist(beta_distribution_a3_b9, density = True)
y = (((bins5**(3-1))*(1-bins5)**(9-1))/sc.beta(3,9));
plt6 = plt.plot(bins5, y, color = 'r')
plt.show()
```

```python
#4 Two dimensional distribution
#with E1
plt7 = plt.hist2d(x1, y1, bins = 100, density = True)
plt.show()
```

```python
#with E2
plt8 = plt.hist2d(x2, y2, bins = 100, density = True)
plt.show()
```
